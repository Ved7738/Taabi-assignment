```python
import numpy as np 
```

```python
def detect_outliers_all(cleaned_data):     #created a function for detecting outlier for all columns by using interquartile range
    outlier_counts = {}
    for column in cleaned_data.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = cleaned_data[column].quantile(0.25)
        Q3 = cleaned_data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = ((cleaned_data[column] < lower_bound) | (cleaned_data[column] > upper_bound)).sum()
        outlier_counts[column] = outliers
    return outlier_counts
```

```python
def handle_outliers_all(cleaned_data, method="cap"):                  # Function to handle outliers for all numerical columns
    for column in cleaned_data.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = cleaned_data[column].quantile(0.25)
        Q3 = cleaned_data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        if method == "cap":
            # Cap outliers
            cleaned_data[column] = np.clip(cleaned_data[column], lower_bound, upper_bound)
        elif method == "remove":
            # Remove rows with outliers
            cleaned_data = cleaned_data[(cleaned_data[column] >= lower_bound) & (cleaned_data[column] <= upper_bound)]
    return cleaned_data
```

```python
outlier_counts = detect_outliers_all(cleaned_data)
print("Number of outliers in each column:")
print(outlier_counts)
```

```python
cleaned_data2 = handle_outliers_all(cleaned_data, method="cap")
print("Outliers handled successfully!")
```
